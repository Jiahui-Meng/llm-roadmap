# Chunking Demo

Compare fixed-size, overlap, and paragraph chunking for a sample document.


```python
sample_text = """Transformers changed sequence modeling by replacing recurrence with attention.

RAG systems build on LLMs by adding retrieval, chunking, indexing, and grounded generation.

Chunking strategy affects retrieval quality, citation clarity, and context efficiency."""
sample_text
```

```python
def fixed_chunk(text, size=60):
    return [text[i:i+size] for i in range(0, len(text), size)]

def paragraph_chunk(text):
    return [p.strip() for p in text.split("\n\n") if p.strip()]

print('Fixed chunks:')
for c in fixed_chunk(sample_text):
    print('-', c)

print('\nParagraph chunks:')
for c in paragraph_chunk(sample_text):
    print('-', c)
```
